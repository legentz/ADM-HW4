```python
from wordcloud import WordCloud
from utils import preprocessing_nltk
import csv
import matplotlib.pyplot as plt
```

```python
huge_descr = []

# datasets containing all words from descriptions
with open('description-dataset-raw.tsv', 'r') as ds:
    ds_csv = csv.reader(ds, delimiter='\t', quotechar=None)
    
    for elems in ds_csv:
        ad_id = elems[0]
        descr = elems[1]
        descr = preprocessing_nltk(descr)
        huge_descr.extend(descr)
```

```python
print(len(huge_descr))
print(huge_descr[:5])
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=len(huge_descr), background_color="white").generate(' '.join(huge_descr))
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
